This notebook contains simple visualization of live tracking data scraped from [NFL next gen stat website](https://nextgenstats.nfl.com) 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
#load data
data = pd.read_csv('output_file/chiefs-at-ravens-2020-reg-3.csv.gz', compression='gzip')
```

```python
data.head()
```

```python
data['data.home_player_x_coord'] = data['data.home_player_x_coord'].apply(lambda x: ast.literal_eval(str(x)))
data['data.home_player_y_coord'] = data['data.home_player_y_coord'].apply(lambda x: ast.literal_eval(str(x)))

data['data.away_player_x_coord'] = data['data.away_player_x_coord'].apply(lambda x: ast.literal_eval(str(x)))
data['data.away_player_y_coord'] = data['data.away_player_y_coord'].apply(lambda x: ast.literal_eval(str(x)))

data['data.ball_x'] = data['data.ball_x'].apply(lambda x: ast.literal_eval(str(x)))
data['data.ball_y'] = data['data.ball_y'].apply(lambda x: ast.literal_eval(str(x)))
```

The code in below cell select a random play and extract the player and ball coordinate data from it. Since each play contains multiple players and balls, the code below tidy up the data having a row for each player by ultilizing `pd.explode` 

```python
play_df = data[data.play_id == 1455]

home_df = play_df[['data.home_player_x_coord','data.home_player_y_coord','data.real_time']].set_index('data.real_time').apply(pd.Series.explode).reset_index().astype(float)
away_df = play_df[['data.away_player_x_coord','data.away_player_y_coord','data.real_time']].set_index('data.real_time').apply(pd.Series.explode).reset_index().astype(float)

ball_df = play_df[['data.ball_x','data.ball_y','data.real_time']].set_index('data.real_time').apply(pd.Series.explode).reset_index().astype(float)

```

```python
home_df.head()
```

The field dimension in the live tracking data is defined by bottom left corner of field as (30,30) and 1 yards equal to 6 coordinate in both direction. As the dimension of the whole field is `120 x 53 1/3` yards (including both endzone), the top right hand corner has a coordinate of (750,350). Also, the data include players and ball of the field, therefore in order to plot a visulaztion of play they has to be excluded.

```python
home_df = home_df[(home_df['data.home_player_x_coord']> 30)&(home_df['data.home_player_x_coord']< 750)&(home_df['data.home_player_x_coord'].astype(float)> 35)&(home_df['data.home_player_y_coord']< 350)]
away_df = away_df[(away_df['data.away_player_x_coord']> 30)&(away_df['data.away_player_x_coord']< 750)&(away_df['data.away_player_y_coord'].astype(float)> 35)&(away_df['data.away_player_y_coord']< 350)]
```

```python
ball_df = ball_df[(ball_df['data.ball_x']> 30)&(ball_df['data.ball_x']< 750)&(ball_df['data.ball_y'].astype(float)> 35)&(ball_df['data.ball_y']< 350)]
```

The player data in NFL live feed is in 10 Hz (10 changes in a second), but for some reason the ball is only in 1 Hz (move once a second). As a result, the data has to be smoothen, and the live feed smooth it by linear interpolate ball coordinate which is also be done here. Also, in order to give a better visulazation, ball coordinate is assumed to be unchanged prior the snap.

```python
snap_time = play_df[play_df['data.event'] == 'Snap']['data.real_time'].values[0]
ball_first_time_array = ball_df.groupby(['data.ball_x','data.ball_y']).first().reset_index()['data.real_time'].values
ball_df.loc[~ball_df['data.real_time'].isin(ball_first_time_array ) & (ball_df['data.real_time'] > snap_time-0.5),['data.ball_x','data.ball_y']] = np.nan
ball_df = ball_df.interpolate(method='slinear').ffill()
```

```python
%%capture
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
from IPython.display import HTML

ylim=(40,85)
xlim=(0,50)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(xlim=xlim, ylim=ylim)
points1, = ax.plot([], [],'.',alpha = 1, markersize =10,color='#241773')
points2, = ax.plot([], [],'.',alpha = 1, markersize =10,color='#E31837')
points3, = ax.plot([], [],'d',alpha = 1, markersize =10,color='brown')

def animate(i):
    time = home_df['data.real_time'].unique()[i]
    if(i > 10):
        ball_time = home_df['data.real_time'].unique()[i-10]
    else:
        ball_time = home_df['data.real_time'].unique()[i]
    
    home_x = home_df[home_df['data.real_time'] == time]['data.home_player_x_coord']
    home_y = home_df[home_df['data.real_time'] == time]['data.home_player_y_coord']

    away_x = away_df[away_df['data.real_time'] == time]['data.away_player_x_coord']
    away_y = away_df[away_df['data.real_time'] == time]['data.away_player_y_coord']
    
    ball_x = ball_df[ball_df['data.real_time'] == ball_time]['data.ball_x']
    ball_y = ball_df[ball_df['data.real_time'] == ball_time]['data.ball_y']
    
    
    home_player_coordinate = pd.DataFrame({'x':home_x,'y':home_y})
    home_player_coordinate = home_player_coordinate[(home_player_coordinate.y <= 350) & (home_player_coordinate.y >= 30)]
    
    away_player_coordinate = pd.DataFrame({'x':away_x,'y':away_y})
    away_player_coordinate = away_player_coordinate[(away_player_coordinate.y <= 350) & (away_player_coordinate.y >= 30)]
    
    points1.set_data((home_player_coordinate['y']-28)/6,(home_player_coordinate['x']-90)/6)
    points2.set_data((away_player_coordinate['y']-28)/6,(away_player_coordinate['x']-90)/6)
    points3.set_data((ball_y-28)/6,(ball_x-90)/6)
    return points1,points2,points3,

anim = animation.FuncAnimation(fig, animate,
                               frames=len(home_df['data.real_time'].unique()), interval=100, blit=True)
```

Then we can plot it as animation
